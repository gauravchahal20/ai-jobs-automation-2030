```python
import os
from pathlib import Path
import pandas as pd
import numpy as np

BASE_DIR = Path().resolve().parent.parent

DATA_PATH = BASE_DIR / "data" / "raw" / "ai_automation_risk_dataset.csv"

df = pd.read_csv(
    DATA_PATH,
    sep=",",
    quotechar='"',
    engine='python',
    encoding='utf-8'
)

print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
print("\nMissing values:\n", df.isnull().sum())

df = df.fillna(df.median(numeric_only=True))
```

```python
df.columns = df.columns.str.lower().str.strip()

df.columns
```

```python
df = df.drop_duplicates()
```

```python
df = pd.get_dummies(df, columns=[
    'industry',
    'education_level',
    'ai_tool_availability'
])
```

```python
df = df.drop(columns=['job_role'])
```

```python
df = df.rename(columns={'automation_risk_score': 'automation_risk'})
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
print("\nFinal Shape:", df.shape)
print(df.head())
print(df.describe())
```

```python
SAVE_PATH = BASE_DIR / "data" / "processed" / "cleaned_data.csv"

df.to_csv(SAVE_PATH, index=False)

print("✅ Preprocessing complete!")
```
