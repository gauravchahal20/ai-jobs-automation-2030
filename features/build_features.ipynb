```python
BASE_DIR = Path().resolve().parent

DATA_PATH = BASE_DIR / "data" / "processed" / "cleaned_data.csv"

df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
print(df.columns)
```

```python
df['automation_pressure'] = (
    0.4 * df['task_repetition_level'] +
    0.4 * df['percent_tasks_automatable'] +
    0.2 * df['ai_dependency_future']
)

df['human_resistance'] = (
    0.4 * df['creativity_requirement'] +
    0.3 * df['social_interaction_level'] +
    0.3 * df['communication_requirement']
)

df['skill_index'] = (
    0.5 * df['analytical_complexity'] +
    0.5 * df['skill_complexity_score']
)

df['ai_readiness'] = (
    0.5 * df['ai_tool_maturity_score'] +
    0.5 * df['ai_dependency_current']
)
```

```python
df['experience_inverse'] = 1 / (df['experience_required_years'] + 1)

df['salary_scaled'] = df['avg_salary_usd'] / 100000
```

```python
# Normalize demand first
df['job_demand_scaled'] = df['job_demand_index'] / df['job_demand_index'].max()

df['job_stability'] = (
    0.5 * df['job_growth_rate'] +
    0.5 * df['job_demand_scaled']
)
```

```python
noise = np.random.normal(0, 0.02, len(df))

df['automation_risk'] = (
    0.30 * df['percent_tasks_automatable'] +
    0.20 * df['task_repetition_level'] +
    0.20 * df['ai_dependency_future'] +
    0.10 * (1 - df['creativity_requirement']) +
    0.10 * (1 - df['social_interaction_level']) +
    0.10 * (1 - df['salary_scaled']) +
    noise
)

df['automation_risk'] = df['automation_risk'].clip(0, 1)
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
SAVE_PATH = BASE_DIR / "data" / "processed" / "featured_data.csv"

df.to_csv(SAVE_PATH, index=False)

print("✅ Feature Engineering Done")
```
